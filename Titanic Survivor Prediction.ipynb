```python
data =pd.read_csv("Train.csv")
```

```python
data.columns
```

```python
columns_to_drop = ["embarked","name","home.dest","cabin","body","ticket","boat"]
data_clean = data.drop(columns_to_drop,axis=1)
Test_clean = Test.drop(columns_to_drop,axis=1)
data_clean.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data_clean["sex"] = le.fit_transform(data_clean["sex"])
Test_clean["sex"] = le.fit_transform(Test_clean["sex"])
data_clean.head()
```

```python
data_clean = data_clean.fillna(data_clean["age"].mean())
Test_clean = Test_clean.fillna(Test_clean["age"].mean())
```

```python
data_clean
```

```python
Test_clean
```

```python
input_cols = ['pclass',"sex","age","sibsp","parch","fare"]
output_cols = ["survived"]

X = data_clean[input_cols]
Y = data_clean[output_cols]

print(X.shape,Y.shape)
print(type(X))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
sk_tree = DecisionTreeClassifier(criterion='entropy',max_depth=5)
sk_tree.fit(X,Y)
sk_tree.score(X,Y)
```

```python
pred=sk_tree.predict(Test_clean)
```

```python
pred=pd.DataFrame(pred)
```

```python
pred.columns=["survived"]
```

```python
pred.insert(0, "Id",np.arange(0,300) , True) 
pred.to_csv("predictions.csv",index=False)
```
